```python
import numpy as np
from selection import boltzmann_selection

# Example usage:

max_generation = 500
generation = 1
k = K                                  # Number of desired selected individuals.
T_0 = 20                                # Initial temperature.
T_offset = 5                          # Offset value for temperature exponential function.
reference_generation = max_generation   # Reference generation where 0 < (T_0 - T_offset) = 0.015 = deltaT is desired.
deltaT = 0.015                          # 0 < (T_0 - T_offset) = deltaT at reference generation.
m = - (np.log(deltaT) - np.log(T_0 - T_offset)) / reference_generation

new_boltzmann_selection = boltzmann_selection(poblacion_inicial,k,generation,T_0,T_offset,m)

new_boltzmann_selection
```
