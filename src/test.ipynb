```python
from population_init import generate
from character import CharacterType
#from src.stats import random_stats

N = 20
maxStatsValue = 150
type = CharacterType.ARCHER

#equipment_data = random_stats(N, maxStatsValue)
poblacion_inicial = generate(N,maxStatsValue,type)

poblacion_inicial
```

#### Seleccion de k indiduos
A*(método1) + (1-A)*(método2)

```python
import numpy as np
#Dado un K
#Si quiero un 40% de la selección con metodo1
K = 10
m1_perc = 0.4
N1 = int(np.round(K * m1_perc))
N2 = K - N1
N1
```

```python
#Añado las columnas de desempeño relativo y acumulado
from selection import add_relative_acumulate
poblacion_rel_acu = add_relative_acumulate(poblacion_inicial)
```

```python
from selection import elitist_selection
nueva_poblacion_e = elitist_selection(poblacion_rel_acu,N1)
nueva_poblacion_e
```

```python
from selection import roulette_wheel_selection
nueva_poblacion_rul = roulette_wheel_selection(poblacion_rel_acu, N2)
nueva_poblacion_rul
```

```python
import pandas as pd
selection_total = pd.concat([nueva_poblacion_e,nueva_poblacion_rul])
selection_total
```

```python
from genes import  encode_genes

cromosomas = encode_genes(selection_total)
cromosomas
```

```python
from crossover import single_point_crossover
new_individuals = single_point_crossover(cromosomas)
new_individuals
```

```python
from genes import decode_genes
cromosomas_decoded = decode_genes(new_individuals)
cromosomas_decoded
```

```python
from crossover import valid_offspring
offspings, offspring_ok = valid_offspring(cromosomas_decoded)

print(f"Cantidad de hijos validos: {offspings}")
offspring_ok
```
